```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
import re
import sys
```

```python
def Get_soup_bs4(url):
    requete = requests.get(url)
    page = requete.content
    soup = BeautifulSoup(page, 'html.parser')
    return soup 
```

```python
def Get_soup_driver1(url):
    driver = webdriver.Firefox()    
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    time.sleep(2)
    return soup
```

```python
def crawler_scrape_data (url_mere, name_file, nbr_page_a_srapper):
    PATH ="Data/"
    All_texte = []
    driver = webdriver.Firefox()    
    driver.get(url_mere)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    time.sleep(3)
    i = 0
    next_page = '//*[@id="pnnext"]/span[2]'
    page_down = "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"

    for i in range(0, nbr_page_a_srapper):
        list_url = []
        try:
                    driver.execute_script(page_down)
                    time.sleep(4)
                    driver.find_element_by_xpath(next_page).click()
                    
                    driver.refresh()
                    soup = BeautifulSoup(driver.page_source, 'html.parser')
   
                    for srg in  soup.find_all('div', {'class': 'srg'}):
                        for le_g in srg.find_all('div', {'class': 'g'}) :
                            for le_rc in le_g.find('div', {'class': 'rc'}):
                                url = le_g.find('div', {'class': 'r'}).find('a').get('href')
                            list_url.append(url)
                    
                    for url_son in list_url:
                        print('url : ', url_son)
                        soup_ = Get_soup_bs4(url_son)
                        for texte in soup_.find_all('p'):
                            All_texte.append(texte.text)

                        with open(PATH + name_file +'.txt', 'w', encoding='utf8') as f_file:
                                    f_file.write(str(All_texte))

                    
                    
                    time.sleep(1)
        except:
                    pass

    driver.quit()
```

```python
if __name__ =="__main__":
    nbr_page_a_scrap = 35
    liste_expression = ['hotel','informatique','science','technologie']
    for querry in liste_expression :
        url = "http://www.google.com/search?q=" + querry
        crawler_scrape_data(url, querry, nbr_page_a_scrap) 
        print('finshed')
```
